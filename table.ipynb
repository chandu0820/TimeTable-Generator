```python

import matplotlib.pylab as pl
def plot(table,secname):
    ny = len(table)
    nx = len(table[0])
    pl.figure("Section "+ secname)
    tb = pl.table(cellText=table, loc=(0,0), cellLoc='center')
    tc = tb.get_child_artists()
    for cell in tc: 
        cell.set_height(1/ny)
        cell.set_width(1/nx)
    ax = pl.gca()
    ax.set_xticks([])
    ax.set_yticks([])

    pl.show()

```
